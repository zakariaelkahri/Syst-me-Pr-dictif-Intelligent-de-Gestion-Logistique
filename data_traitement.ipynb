```python
from app.sparksession import spark

df = spark.read.csv("./data/data_cleaned/part-00000-ebc0140e-389c-4ae9-990d-99a909412869-c000.csv", header=True, inferSchema=True)

df.printSchema()
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.stat import Correlation

# ðŸ‘‰ Select only numeric columns
num_cols = ["Days for shipment (scheduled)", "Late_delivery_risk", "Category Id"]

# Convert columns into a single "features" vector column
assembler = VectorAssembler(inputCols=num_cols, outputCol="features")
df_vector = assembler.transform(df).select("features")

# Compute correlation matrix
corr_matrix = Correlation.corr(df_vector, "features", "pearson").collect()[0][0]

print("âœ… Matrice de corrÃ©lation :")
print(corr_matrix)
```
